## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Bins are 0, 10, 14, 19, 24, 29, 34, 39, 100. 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create Age Ranges
Age_Groups = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
```

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=Age_Groups)
AgeRangedf = purchase_data.set_index("Age Ranges")
AgeRangedf

```

```python
#Find Total People in Age Range
AgeRangedffiltered = AgeRangedf.drop_duplicates(subset= 'SN', keep='first', inplace=False)
pop_under10 = AgeRangedffiltered["Purchase ID"].loc["<10"].count()
pop_10_14 = AgeRangedffiltered["Purchase ID"].loc["10-14"].count()
pop_15_19 = AgeRangedffiltered["Purchase ID"].loc["15-19"].count()
pop_20_24 = AgeRangedffiltered["Purchase ID"].loc["20-24"].count()
pop_25_29 = AgeRangedffiltered["Purchase ID"].loc["25-29"].count()
pop_30_34 = AgeRangedffiltered["Purchase ID"].loc["30-34"].count()
pop_35_39 = AgeRangedffiltered["Purchase ID"].loc["35-39"].count()
pop_over40 = AgeRangedffiltered["Purchase ID"].loc["40+"].count()
Total_Players = AgeRangedffiltered["Purchase ID"].count()
Total_Count = ({"Total_Count":[pop_under10, pop_10_14, pop_15_19, pop_20_24, pop_25_29, pop_30_34, pop_35_39, pop_over40]})
Total_Countdf = pd.DataFrame(Total_Count, Age_Groups)
Total_Countdf
```

```python
perc_under10 = '{:.2%}'.format(pop_under10/Total_Players)
perc_10_14 = '{:.2%}'.format(pop_10_14/Total_Players)
perc_15_19 = '{:.2%}'.format(pop_15_19/Total_Players)
perc_20_24 = '{:.2%}'.format(pop_20_24/Total_Players)
perc_25_29 = '{:.2%}'.format(pop_25_29/Total_Players)
perc_30_34 = '{:.2%}'.format(pop_30_34/Total_Players)
perc_35_39 = '{:.2%}'.format(pop_35_39/Total_Players)
perc_over40 = '{:.2%}'.format(pop_over40/Total_Players)
PercofPlayers = ({"Percent_of_Players": [perc_under10,perc_10_14,perc_15_19,perc_20_24,perc_25_29,perc_30_34,perc_35_39,perc_over40]})
PercentofPlayersdf = pd.DataFrame(PercofPlayers, Age_Groups)
PercentofPlayersdf
```

```python
#Combine Data Frames
AgeDemographicsdf = pd.concat([Total_Countdf, PercentofPlayersdf], axis=1, join='outer')
AgeDemographicsdf
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Popitems = purchase_data.groupby(['Item Name'])
Popitemsdf = pd.DataFrame(Popitems)
Popitemsdf
```

```python
UniqueItemCounts = purchase_data['Item Name'].value_counts()
UniqueItemCountsdf = pd.DataFrame(UniqueItemCounts)
UniqueItemCountsdf
```

```python
ItemID = purchase_data[['Item ID', 'Item Name']]
ItemID = ItemID.set_index('Item Name')
ItemID
```

```python
Popitemssum = purchase_data.groupby(['Item Name']).sum()
Popitemssumdf = Popitemssum.sort_values(['Price'], ascending=False)
Popitemssumdf = Popitemssumdf.drop(columns=['Purchase ID', 'Age', 'Item ID'])
Popitemssumdf['Purchase Count']= UniqueItemCounts
Popitemssumdf['Item Price']= Popitemssumdf['Price']/Popitemssumdf['Purchase Count']
Popitemssumdf=Popitemssumdf.join(ItemID, on='Item Name')
Popitemssumdf=Popitemssumdf.drop_duplicates(subset= 'Item Price', keep='first', inplace=False)
Popitemssumdf=Popitemssumdf.style.format({'Price' : "${:,.2f}",'Item Price':"${:,.2f}"})
Popitemssumdf
```
