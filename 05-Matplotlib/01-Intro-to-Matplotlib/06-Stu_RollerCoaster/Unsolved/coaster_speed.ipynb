```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
DangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun= [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
Time= np.arange(0,130,10)
```

```python
Time
```

```python
plt.plot(Time, DangerDrop, marker='o',color='blue',label='Danger Drop')
plt.plot(Time, RailGun, marker='^',color='red',label='Rail Gun')
plt.legend(loc="best")
plt.title("Coaster Speed Over Time")
plt.xlabel("Time")
plt.ylabel("Speed")
plt.grid()
plt.xlim(0,120)
plt.ylim(5,95)
plt.show()
```
